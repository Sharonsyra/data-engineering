```python
import json
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the JSON file
with open('/Users/waithira/Projects/data-engineering/data/people_size_by_pob.json') as f:
    data = json.load(f)
```

```python
# Display the JSON content
print(json.dumps(data, indent=4))
```

```python
# If the data is suitable for tabular representation
df = pd.DataFrame([data])
display(df)
```

```python
# If the JSON contains numerical data, you can also plot it
# Example plot (customize based on your data structure)
df.plot(kind='bar')
plt.title('Summary Data')
plt.xlabel('Metrics')
plt.ylabel('Values')
plt.show()
```
